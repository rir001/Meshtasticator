```python
import numpy as np
import yaml
```

```python
base = {
    'antennaGain': 1.0,
    'hopLimit': 3,
    'isClientMute': False,
    'isRepeater': False,
    'isRouter': False,
    'neighborInfo': True,
    'x': 0,
    'y': 0,
    'z': 1
}
```

```python
x_n = 3
y_n = 3

x_step = 1000
y_step = 1000

data = {}
count = 0
for x_ in range(x_n):
    for y_ in range(y_n):
        data[count] = base.copy()
        data[count]["x"] = x_step * x_ - x_n * x_step // 2
        data[count]["y"] = y_step * y_ - y_n * y_step // 2
        count += 1
```

```python
yaml_output = yaml.dump(data, default_flow_style=False)

with open('out/nodeConfig.yaml', 'w') as f:
    f.write(yaml_output)
```
